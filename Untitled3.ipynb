```python
import numpy as np

# Define the simulation parameters
n = 50              # size of the grid (n x n cells)
probTree = 0.6      # probabilitya that a cell is initialized as a non-burning tree
probBurning = 0.01  # probability that a non-burning tree is selected and initialized as a burning tree
pIgnite = 0.01      # probability that a non-burning tree becomes a burning tree if at least one of its neighbors is a burning tree
pLightning = 0.001  # probability that a random non-burning tree is selected and initialized as a burning tree due to a lightning strike

# Initialize the grid
grid = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        if np.random.random() < probTree:
            grid[i, j] = 1
        if grid[i, j] == 1 and np.random.random() < probBurning:
            grid[i, j] = 2

# Define a helper function to get the neighbors of a given cell
def get_neighbors(i, j):
    """
    Returns a list of the (i, j) coordinates of the neighbors of the cell (i, j).
    """
    neighbors = []
    for di in [-1, 0, 1]:
        for dj in [-1, 0, 1]:
            if di == 0 and dj == 0:
                continue
            ni = i + di
            nj = j + dj
            if 0 <= ni < n and 0 <= nj < n:
                neighbors.append((ni, nj))
    return neighbors

# Run the simulation for a fixed number of time steps
num_steps = 100
for t in range(num_steps):
    # Check for lightning strikes
    for i in range(n):
        for j in range(n):
            if grid[i, j] == 1 and np.random.random() < pLightning:
                grid[i, j] = 2
    
    # Update the state of each cell based on the states of its neighbors
    new_grid = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            if grid[i, j] == 0:
                new_grid[i, j] = 0
            elif grid[i, j] == 2:
                new_grid[i, j] = 0
            else:
                num_burning_neighbors = 0
                for neighbor in get_neighbors(i, j):
                    if grid[neighbor] == 2:
                        num_burning_neighbors += 1
                if num_burning_neighbors > 0 and np.random.random() < pIgnite:
                    new_grid[i, j] = 2
                else:
                    new_grid[i, j] = 1
    
    # Update the grid for the next time step
    grid = new_grid

    # Visualize the grid at each time step (optional)
    # import matplotlib.pyplot as plt
    # plt.imshow(grid, cmap='hot', interpolation='nearest')
    # plt.show()
```
