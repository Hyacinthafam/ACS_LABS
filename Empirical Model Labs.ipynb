```python
import numpy as np
import matplotlib.pyplot as plt
#np.polyfit(x-coordinates, y-coordinates, n)

x = np.array([-3, 2, -1, 4, 0])
y = np.array([20, 20, -10, 30, 0])
fitCubic = np.polyfit(x, y , 3)
print (fitCubic) #returns the list of cubic polynomial coefficients

xp = np.linspace(-4, 6, 100)
polyValues = np.polyval(fitCubic, xp)

%matplotlib inline

plt.plot(x, y, '.') # the original x, y
plt.plot(xp, polyValues, 'g--') # the estimated values
plt.show()





```
