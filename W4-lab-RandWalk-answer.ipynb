```python
from random import *
from math import *

def generateCordinate(prob, XY):
    # store movement in a list
    directions = ["N", "NE", "E", "SE", "S", "SW", "W", "NW"]
    directionMove = directions[prob];
    
    if directionMove == "N":
        return [XY[0]-1, XY[1]]
    elif directionMove == "NE":
        return [XY[0]-1, XY[1]+1]
    elif directionMove == "E":
        return [XY[0], XY[1]+1]
    elif directionMove == "SE":
        return [XY[0]+1, XY[1]+1]
    elif directionMove == "S":
        return [XY[0]+1, XY[1]]
    elif directionMove == "SW":
        return [XY[0]+1, XY[1]-1]
    elif directionMove == "W":
        return [XY[0], XY[1]-1]
    elif directionMove == "NW":
        return [XY[0]-1, XY[1]-1]    
```

```python
def RandomWalk(n):
    # determine points of path
    x = 0
    y = 0
    xLst = [x] 
    yLst = [y] 
    for i in range(n):
        x, y = generateCordinate(randrange(8), [x, y])
        xLst.append(x)
        yLst.append(y)
        
    return xLst, yLst
```

```python
def RandomWalkDistance(xLst, yLst):
    distLst = []
    tList = []
    for i in range(len(xLst)):
        dist = sqrt((xLst[i]**2) + (yLst[i]**2))
        distLst.append(dist)
        tList.append(i)
    
    return tList, distLst
        
```

```python
# instead of matplotlib inline. It turns the notebook interactive
%matplotlib notebook  

import matplotlib.pyplot as plt
from matplotlib import animation 

x, y = RandomWalk(50)


fig = plt.figure()                           
ax = plt.axes(xlim=(min(x), max(x)), ylim=(min(y),max(y)))
ax.plot([], [], "r-", lw=1)  

#def init(): 
    #ax.plot([], [], "r-", lw=1) 

def animate(i):
    ax.plot(x[:i], y[:i],"r-", lw=1)  
    ax.plot(x[:i], y[:i], "ro")

anim = animation.FuncAnimation(fig, animate, frames=range(len(x)), interval=500, blit=True)
plt.show()
```

```python
import numpy as np

def averageDistanceCovered(numTests):
    allTime = []
    allDist = []
    for i in range(numTests):
        x, y = RandomWalk(50)
        tList, distLst = RandomWalkDistance(x, y)
        allTime.append(tList)
        allDist.append(distLst)
    return np.array(allTime), np.array(allDist)

allTime, allDist = averageDistanceCovered(20)

avgTime = np.mean(allTime, axis=0)
avgDiste = np.mean(allDist, axis=0)

fig = plt.figure()  
plt.plot(avgTime, avgDiste, "go")
plt.xlabel('n steps')
plt.ylabel('distance')
plt.show()



```
