```python
import numpy as np
import matplotlib.pyplot as plt

# Initial population of the animal
initial_population = 15000

# Continuous growth rate of 2% per year
growth_rate = 0.02

# Time step size of 0.25 years
time_step = 0.25

# Simulation length of 20 years
simulation_length = 20

# Loop through the simulation length
time = 0
population = initial_population
times = []
populations = []
actual_populations = []
relative_errors = []
while time <= simulation_length:
    # Calculate the new population using exponential growth
    actual_population = initial_population * np.exp(growth_rate * time)
    # Calculate the relative error
    relative_error = (population - actual_population) / actual_population
    # Add the new time and population to the arrays
    times.append(time)
    populations.append(population)
    actual_populations.append(actual_population)
    relative_errors.append(relative_error)
    # Increase the time by the time step
    time += time_step

# Plot the population and actual population against time on the same plot
plt.plot(times, populations, label="Simulated Population")
plt.plot(times, actual_populations, label="Actual Population")
plt.xlabel("Time (years)")
plt.ylabel("Population")
plt.legend()
plt.show()

# Determine the population at the end of 20 years
print("Time (years)\tSimulated Population\tActual Population\tRelative Error")
print(f"{simulation_length:.2f}\t{population:.2f}\t\t{actual_population:.2f}\t\t{relative_error:.2%}")

# Determine the growth and the population at the end of the first three time steps
print("Time (years)\tSimulated Population\tActual Population\tRelative Error")
for i in range(3):
    print(f"{times[i]:.2f}\t{populations[i]:.2f}\t\t{actual_populations[i]:.2f}\t\t{relative_errors[i]:.2%}")
```
