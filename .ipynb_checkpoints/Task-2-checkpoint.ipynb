```python
import numpy as np
import pandas as pd
from dateutil.parser import parse
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
path = "Task-2.csv"
```

```python
df = pd.read_csv(path)
df.head()
```

```python
from statsmodels.tsa.stattools import adfuller
from numpy import log
```

```python
result = adfuller(df.particullate_matter.dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
# P value less than 0.05, therefore no differencing is needed, d = 0
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})


# Original Series
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df.particullate_matter); axes[0].set_title('Original Series')
plot_acf(df.particullate_matter, ax=axes[ 1])

plt.show()

```

```python
#find value of p
#find value of p
# PACF plot of 1st differenced series
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df.particullate_matter); axes[0].set_title('original series')
axes[1].set(ylim=(0,5))
plot_pacf(df.particullate_matter.dropna(), ax=axes[1])

plt.show()
```

```python
#find (q)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df.particullate_matter); axes[0].set_title('original series')
axes[1].set(ylim=(0,1.2))
plot_acf(df.particullate_matter, ax=axes[1])

plt.show()


```

```python
#q value is taken as 1 
```

```python
from statsmodels.tsa.arima.model import ARIMA

# 1,1,2 ARIMA Model
model = ARIMA(df.particullate_matter, order=(1,0,1))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# 1,1,1 ARIMA Model
model = ARIMA(df.particullate_matter, order=(1,0,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Plot residual errors
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

```python
# Actual vs Fitted
#from statsmodels.graphics.tsaplots import plot_predict
# Generate predictions
predictions = model_fit.predict(dynamic=False)

# Plot the predictions
plt.plot(predictions, label='Predictions')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import acf

# Create Training and Test
train = df.particullate_matter[:14055]
test = df.particullate_matter[14055:]
```

```python
from pmdarima.arima import auto_arima

# Find the best ARIMA model using auto_arima
model = auto_arima(train, start_p=1, start_q=1, max_p=3, max_q=3, 
                   m=1, start_P=0, seasonal=False, d=0, D=0, trace=True, 
                   error_action='ignore', suppress_warnings=True, stepwise=True)

# Fit the model to the training data
model.fit(train)

# Forecast
fc, conf_int = model.predict(n_periods=len(test), return_conf_int=True)

# Make as pandas series
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf_int[:, 0], index=test.index)
upper_series = pd.Series(conf_int[:, 1], index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.15)
plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()


```
