```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("filip.csv")
x = data["x"].values
y = data["y"].values
fitCubic = np.polyfit(x, y , 3)
print (fitCubic) #returns the list of cubic polynomial coefficients

xp = np.linspace(0, 6, 100)
polyValues = np.polyval(fitCubic, xp)

%matplotlib inline

plt.plot(x, y, 'o') # the original x, y
plt.plot(xp, polyValues, 'g--') # the estimated values
plt.show()
```
